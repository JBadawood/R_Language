### Mean & Variance (MTTF)

image.png

Low Breakdowns

```r
# Define shape and scale parameters
shape_param <- 1.1
scale_param <- 20

# Calculate Expected Value
expected_value <- scale_param * gamma(1 + (1 / shape_param))
print(paste("Expected Value:", expected_value))

# Calculate Variance
variance <- (scale_param^2) * (gamma(1 + (2 / shape_param)) - (gamma(1 + (1 / shape_param)))^2)
print(paste("Variance:", variance))
```

High Breakdowns

```r
# Define shape and scale parameters
shape_param <- 2.6
scale_param <- 10

# Calculate Expected Value
expected_value <- scale_param * gamma(1 + (1 / shape_param))
print(paste("Expected Value:", expected_value))

# Calculate Variance
variance <- (scale_param^2) * (gamma(1 + (2 / shape_param)) - (gamma(1 + (1 / shape_param)))^2)
print(paste("Variance:", variance))
```

### Mean & Variance (MTTR)

Low Breakdowns

```r
# Define the parameters of the lognormal distribution
meanlog <- 1.39  # Example mean of the log-transformed data
sdlog <- 0.46   # Example standard deviation of the log-transformed data

# Calculate the Expected Value (Mean)
expected_value <- exp(meanlog + (sdlog^2) / 2)
print(paste("Expected Value:", expected_value))

# Calculate the Variance
variance <- exp(2 * meanlog + sdlog^2) * (exp(sdlog^2) - 1)
print(paste("Variance:", variance))
```

High Breakdowns

```r
# Define the parameters of the lognormal distribution
meanlog <- 3.9  # Example mean of the log-transformed data
sdlog <- 0.3   # Example standard deviation of the log-transformed data

# Calculate the Expected Value (Mean)
expected_value <- exp(meanlog + (sdlog^2) / 2)
print(paste("Expected Value:", expected_value))

# Calculate the Variance
variance <- exp(2 * meanlog + sdlog^2) * (exp(sdlog^2) - 1)
print(paste("Variance:", variance))
```
